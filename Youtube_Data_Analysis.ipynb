```python
title = {}
for i in Cat['items'].values:
  j = i['snippet']['title']
  title[i['id']]= [j]
  
```

```python

genre = pd.DataFrame({'category_id':title.keys(),'Genre':title.values()})
```

```python
genre['Genre']=[i[0] for i in genre['Genre'].values]
```

```python
genre
```

```python
genre['category_id']=pd.to_numeric(genre['category_id'])
```

```python


data = pd.merge( US_data,genre,on='category_id',how='inner')
```

Here we are extracting trending day, trending,month, trending year from the column

```python
day = []
month = []
year = []
for i in data['trending_date'].values:
  day.append(i[3:5])
  month.append(i[6:])
  year.append(i[:2])
```

```python
data['t_day']=day
data['t_month']=month
data['t_year'] = year
```

```python
con = widgets.Dropdown(
    options=['views', 'likes', 'dislikes','comment_count'],
    value='views',
    description='Select option:',
    disabled=False,
)

def plotting(x_value):
  plt.figure(figsize = (25,13))
  sns.barplot(y = x_value,x = 'Genre',data = data)
  plt.ylabel(x_value.upper(),fontsize=18)
  plt.xlabel('Genre'.upper(),fontsize=18)
  plt.xticks(rotation='vertical')
  plt.show()
```

```python
widgets.interact(plotting,x_value = con)
```

```python
# Graphical Representation of Videos based on trending days
```

Here we are Aggregating the data on title column due to rerepetetive data of views and other columns

```python
trend_data = data[['title','t_day','t_month','t_year']].groupby(by='title').count()
```

```python
trend_data2 = data[['title','Genre','views','likes','dislikes','comment_count']].groupby(by='title').max()
```

```python
trend_data2
```

```python
main_trend = pd.merge(trend_data,trend_data2,how = 'inner',on='title')
```

```python
#Here we finally filtered our data according to trend
main_trend.sort_values(by='t_day',ascending=False)
```

```python

# Lets make a beautiful GUI for the representation
```

```python
# Graph between Genre and trending days
plt.figure(figsize = (25,13))
sns.boxplot(x = 'Genre',y = 't_day',data = main_trend)
plt.ylabel('Trending Days',fontsize=18)
plt.xlabel('Genre',fontsize=18)
plt.xticks(rotation='vertical')
plt.show()
```

graph between trending days and genre clearly indicating that the shows are getting more number of trend days as compared to nepotism and activity and here black dots that we are seeing here it is indicating the outliers.

```python

con2 = widgets.Dropdown(
    options=list(main_trend['Genre'].unique()),
    value='Shows',
    description='X axis:',
    disabled=False,
)

def plotting(x_value):
  temp_d = main_trend[main_trend['Genre']==x_value]
  # print(temp_d.index)
  con3 = widgets.Dropdown(
    options=list(temp_d.index),
    value=None,
    description='Y axis:',
    disabled=False,
  )
  
  def mainplot(y_value):
    try:
      temp_t = temp_d[temp_d.index== y_value]
      plt.figure(figsize = (20,10))
      # declaring data
      datap = [temp_t['likes'].values[0],temp_t['dislikes'].values[0]]
      keys = ['Likes','Dislikes']
        
      # define Seaborn color palette to use
      palette_color = sns.color_palette('bright')
        
      # plotting data on chart
      plt.pie(datap, labels=keys, colors=palette_color, autopct='%.0f%%')
        
      # displaying chart
      plt.show()
      # print(temp_t.head())
      # sns.lineplot(x = 't_day',y = 'views',data = temp_t,marker='*')
      # plt.show()
      # # plt.ylabel(x_value.upper(),fontsize=18)
      # # plt.xlabel(y_value.upper(),fontsize=18)
      # # plt.xticks(rotation='70')
    except:
      pass
  widgets.interact(mainplot,y_value = con3)
```
